```python
import numpy as np
```

```python
# Initial values
w = np.array([0, 0])
b = 0

# data points
x1 = np.array([-1, 1])
x2 = np.array([0, -1])
x3 = np.array([10, 1])

# Class
y1 = 1
y2 = -1
y3 = 1
```

```python
# sign function
def sign(w, x, b):
    if w.dot(x) + b > 0:
        return 1
    else:
        if w.dot(x) + b == 0:
            return 0
        else:
            return -1
    
# indicator function
def ind(s, y):
    if s != y:
        return 1
    else:
        return 0
```

```python
x = [x1, x2, x3]
y = [y1, y2, y3]
```

```python
sum = 1
while sum > 0:
    sum = 0
    for i in range(len(x)):
        print("\nData point number {}".format(i+1))
        print("x_{} =".format(i+1), x[i])
        print("")
        print("   |")
        print("   V", "\n")
        print("w =", w, "and b =", b)
        print("")
        print("   |")
        print("   V", "\n")
        print("sign function")
        print("")
        print("   |")
        print("   V", "\n")
        print(sign(w, x[i], b))
        print("")
        print("   |")
        print("   V", "\n")
        print("{} indicator function {}".format(sign(w, x[i], b), y[i]))
        print("")
        print("   |")
        print("   V", "\n")
        print(ind(sign(w, x[i], b), y[i]))
        print("")
        sum = sum + ind(sign(w, x[i], b), y[i])
        if ind(sign(w, x[i], b), y[i]) == 1:
            print("Update w and b")
            w = w + y[i]*x[i]
            b = b + y[i]
            print("   |")
            print("   V", "\n")
            print("w =",w, "and b =", b)
    print("\nNumber of misclassifications",sum)
```
